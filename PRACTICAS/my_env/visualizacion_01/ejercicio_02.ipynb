```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
# Carga del DataSet
wine = datasets.load_wine()
wine
```

```python
# Creamos el DataFrame
# df = pd.DataFrame(wine.data)

# Linea corregida
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df
```

```python
df['target'] = wine.target
df
```

```python
plt.figure(figsize=(8, 6))

for target in range(3):
    subset = df[df['target'] == target]
    plt.scatter(subset['alcohol'], subset['color_intensity'], label=f'Clase {target}')
    
plt.title('Alcohol vs Color Intensity')
plt.xlabel('Alcohol (% vol)')
plt.ylabel('Color Intensity')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['alcohol'], bins=20, color='purple', edgecolor='k', alpha=0.6)
plt.title('Distribución del alcohol en el vino')
plt.xlabel("Alcohol (% vol)")
plt.ylabel("Frecuencia")
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(df['color_intensity'], bins=20, color='purple', edgecolor='k', alpha=0.6)
plt.title('Distribución del alcohol en el vino')
plt.xlabel("Alcohol (% vol)")
plt.ylabel("Frecuencia")
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(df['magnesium'], bins=20, color='purple', edgecolor='k', alpha=0.6)
plt.title('Distribución del alcohol en el vino')
plt.xlabel("Alcohol (% vol)")
plt.ylabel("Frecuencia")
plt.show()
```
