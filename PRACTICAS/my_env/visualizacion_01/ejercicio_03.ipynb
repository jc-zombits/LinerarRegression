```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
california_housing = datasets.fetch_california_housing()
california_housing
```

```python
df = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
df
```

```python
df['target'] = california_housing.target
df
```

```python
plt.figure(figsize=(8, 6))

for target in range(len('target')):
    subset = df[df['target'] == target]
    plt.scatter(subset['HouseAge'], subset['Population'], label=f'{target}')
    
plt.title('Valor Casa')
plt.xlabel('Años de la Casa')
plt.ylabel('Población')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['HouseAge'], bins=20, color='purple', edgecolor='k', alpha=0.6)
plt.title('Casa en Califormia')
plt.xlabel("Años de la casa")
plt.ylabel("Frecuencia")
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(df['Population'], bins=20, color='purple', edgecolor='k', alpha=0.6)
plt.title('Casa en Califormia')
plt.xlabel("Población")
plt.ylabel("Frecuencia")
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(df['MedInc'], bins=20, color='purple', edgecolor='k', alpha=0.6)
plt.title('Casa en Califormia')
plt.xlabel("MedInc")
plt.ylabel("Frecuencia")
plt.show()
```
