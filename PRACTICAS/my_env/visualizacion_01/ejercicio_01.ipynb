```python
# Importamos las bibliotecas necesarias
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
# Carga de los datos iris para este ejemplo
iris = datasets.load_iris()
iris
```

```python
# Convertimos los datos a un DataFrame
# iris.data contiene las características
# iris.feature_names establece los nombres de las columnas como los nombres de las características
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
# Agregamos la columna de especies
df["species"] = iris.target
df
```

```python
# Creamos un gráfico de dispersión

plt.figure(figsize=(8, 6)) # -> Define el tamaño de la figura
species = ['Setosa', 'Versicolor', 'Virginica'] # -> Define un array con el nombre de cada una de las especies

# Iteramos sobre las especies usando **enumerate** para obtener tanto el indice(que representa la especie) como el nombre
for i, species_name in enumerate(species):
    # Filtra el DataFrame para obtener solo las filas que corresponden a la especie actual
    subset = df[df['species'] == i]
    # Creamos un gráfico de dispersión usando las columnas de largo y ancho del sépalo, etiquetando cada especie con un color diferente
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=species_name)
    
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
plt.show()
```

```python
# Creamos un histograma
# bins=20, define cuántos intervalos usar para agrupar los datos
plt.figure(figsize=(8, 6))
plt.hist(df['sepal length (cm)'], bins=20, color='c', edgecolor='k', alpha=0.7)
plt.title('Distribución del largo del Sépalo')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Matriz de dispersión (Pair Plot)
import seaborn as sns

# Cargamos el dataset con nombres de las especies
df['species'] = df['species'].map({0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'})

# Matriz de dispersión que muestra todas las combinaciones de características
# hue='species' coloriza los puntos segun la especie de la flor
sns.pairplot(df, hue='species', markers=['o', 's', 'D'])
plt.show()
```
