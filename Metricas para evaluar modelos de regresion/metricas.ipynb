# Mtricas para Evaluar Modelos de Regresin con Scikit-Learn: R2, MSE, RMSE, MAE, etc.

- Un modelo de regresin es una funcin que nos premite explicar la relacin entre una o multiples variables y una variable objetivo.
- En otras palabras, es una funcin que nos premite predecir un valor en funcin de ciertas variables.
>
> 

```python
# Ejemplo 1:
# ERROR ABSOLUTO MAXIMO (M) - Cul es la diferencia mxima entre el valor predicho y el valor real.

# y: valore verdadero
# : valor predicho

# M(y, ) = max(|y_i - _i|)

from sklearn.metrics import max_error

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
max_error(y_verdadero, y_predicho)
```

```python
# ERROR ABSOLUTO MEDIO (MAE)

# y: valor verdadero
# : valor predicho
# n: tamao de la muestra

# MAE(y, ) = 1/n |y_i, _i|

from sklearn.metrics import mean_absolute_error

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
mean_absolute_error(y_verdadero, y_predicho)
```

```python
# ERROR CUADRATICO MEDIO (MSE)

# # y: valor verdadero
# : valor predicho
# n: tamao de la muestra

# MSE(y, ) = 1/n (y_i - _i)

from sklearn.metrics import mean_squared_error

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
mean_squared_error(y_verdadero, y_predicho)
```

```python
# SUMA DE LOS CUADRADOS DE LOS RESIDUOS (RSS)

# # y: valor verdadero
# : valor predicho
# n: tamao de la muestra

# RSS(y, ) = (y_i - _i)

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
mean_squared_error(y_verdadero, y_predicho)*len(y_predicho)
```

```python
# RAIZ CUADRADA DEL ERROR CUADRATICO MEDIO (RMSE)

# y: valor verdadero
# : valor predicho
# n: tamao de la muestra

# RMSE(y, ) = RaizCuadrada de 1/n (y_i - _i)

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
mean_squared_error(y_verdadero , y_predicho, squared=False)
```

```python
# R (COEFICIENTE DE DETERMINACION)

# y: valor verdadero
# : valor predicho
# n: tamao de la muestra
# yRayaArriba: promedio de los valores verdaderos

# R(y, ) = 1 - (y_i - _i) / (y_i - yRayaArriba)

from sklearn.metrics import r2_score

y_verdadero = [1, 2, 3, 4, 5]
y_predicho = [1, 2, 3, 4, -5]
r2_score(y_verdadero, y_predicho)
```
