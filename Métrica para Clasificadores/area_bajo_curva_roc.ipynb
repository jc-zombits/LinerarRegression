# Métrica esencial para clasificadores: Area Bajo la Curva ROC (Receiver Operating Characteristic)

- Clasificación binaria con clases separables.
- Clase "roja" -> Verdaderos Positivos
- Clase "azul" -> Verdaderos Negativos

> Esta métrica esta definida para clasificadores binarios, es decir, clasificadores que solo pueden discernir entre dos clases.
>
> Debemos tener en cuenta que hay algunas características que son comunes entre las clases. Por lo que para los clasificadores es muy difícil indicar para una cierta instancia en particular a que tipo de clase pertenece.
>
> Por esto, los clasificadores, tienden a cometer diferentes tipos de errores. Estos errores son conocidos como **Falsos Positivos** y **Falsos Negativos**.
>
- **Vamos a darnos un ejemplo en función del dominio de una enfermedad**
>
> Imaginemos que los pacientes de la clase roja son pacientes que sufren alguna enfermedad y la clase azul son personas sanas.
>
    >> Si en este caso, nuestro clasificador clasifica a una persona que tiene una enfermedad como **"Si tiene una enfermedad"**, eso es un **verdadero positivo**.
>
    >> Si en el caso, nuestro clasificador clasifica a una persona sana como **"Si es una persona sana"**, entonces eso sería un **verdadero negativo**.
>
    >> Ahora, si bién nuestro clasificador, clasifica a una persona sana como **"Es una persona enferma"**, entonces tenemos un **falso positivo**.
>
    >> Por el contrario, se clasifica una persona que tiene una enfermedad como **"es una persona sana"**, entonces tenemos un **falso negativo**.
>
> ¿Esto porque es importante para el área bajo la curva roc?
>
    >> esta métrica parte del supuesto de que los clasificadores, pueden definir un umbral.
>

>

>
